# Dim_Order_Mstr

## Cambiamos el formato fecha para aplicarlo a las transformadas

```python
dim_order_mstr = ['Order ID','Order Date'] 
```

```python
nuevo_dim_order = df_nuevo[dim_order_mstr].copy()
```

```python
nuevo_dim_order.to_csv('dim_order_mstr.csv', index=False)
```

```python
df_order = pd.read_csv('./dim_order_mstr.csv', encoding='ISO-8859-1')
df_order
```

```python
df_order['Order Date'] = pd.to_datetime(df_order['Order Date'])
df_order['Year'] = df_order['Order Date'].dt.year
df_order['Month'] = df_order['Order Date'].dt.month
df_order['Day'] = df_order['Order Date'].dt.day
```

```python
df_order['Month_Name'] = df_order['Order Date'].dt.strftime('%B')
```

```python
df_order
```

```python
df_order.columns
```

```python
dim_order_mstr_ampliado = ['Order ID','Order Date','Year', 'Month', 'Day', 'Month_Name'] 
```

```python
nuevo_dim_order_ampliado = df_order[dim_order_mstr_ampliado].copy()
```

```python
nuevo_dim_order_ampliado.to_csv('dim_order_mstr_ampliado.csv', index=False)
```

```python
df_fecha = pd.read_csv('./dim_order_mstr_ampliado.csv', encoding='utf-8', engine='python')
df_fecha
```

Hemos extraido un csv que contiene nuestra order id y nuestra order date a su vez hemos denormalizado la fecha para poder obtener multiples columnas para el análisis.
