```python
with open('Insert_Fact_Sales_mstr.sql', 'w') as archivo_sql:
    for indice, fila in df.iterrows():
        valores = ', '.join([f"'{valor}'" if isinstance(valor, str) else str(valor) for valor in fila])
        comando_sql = f"INSERT INTO fact_sales_mstr (Row_ID, Customer_ID, Product_ID, ship_ID, Category_ID, City_ID, Sales, Quantity, Discount, Profit) VALUES ({valores});\n"
        archivo_sql.write(comando_sql)
```
