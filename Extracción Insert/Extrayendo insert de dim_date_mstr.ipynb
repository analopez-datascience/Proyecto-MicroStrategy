```python
for indice, fila in df.iterrows():
    valores = ', '.join([f"'{valor}'" if isinstance(valor, str) else str(valor) for valor in fila])
    comando_sql = f"INSERT INTO dim_date_mstr (Date_ID, Order Date, Year, Month, Day, Month_Name) VALUES ({valores});"
    print(comando_sql)
```

```python
with open('insert_dim_date_mstr.sql', 'w') as archivo_sql:
    for indice, fila in df.iterrows():
        valores = ', '.join([f"'{valor}'" if isinstance(valor, str) else str(valor) for valor in fila])
        comando_sql = f"INSERT INTO dim_date_mstr (Date_ID, Order Date, Year, Month, Day, Month_Name) VALUES ({valores});\n"
        archivo_sql.write(comando_sql)
```
